## ✅ 어깨 양쪽 평균 column 생성 (center) 

    이미지 너비 기준 (x좌표)

```python
shoulder_df['total'] = shoulder_df.loc[:,33] + shoulder_df.loc[:,36]
shoulder_df['center'] = shoulder_df['total']/2
shoulder_df
```

```python
shoulder_df['left_wrist']=data.loc[:,[45]]
shoulder_df['right_wrist']=data.loc[:,[48]]
shoulder_df
#left_wrist_df=data.loc[:,[45, 46]]
#right_wrist_df=data.loc[:,[48, 49]]
```

# ✅ 왼쪽, 오른쪽 손목 & 어깨 center 비교

    left result  = True 면 왼쪽으로 누운 상태
    right result = True 면 오른쪽으로 누운 상태

```python
shoulder_df['left_result'] = shoulder_df['center'] > shoulder_df['left_wrist']
shoulder_df['right_result'] = shoulder_df['center'] < shoulder_df['right_wrist']
```
